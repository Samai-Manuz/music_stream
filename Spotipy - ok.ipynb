DATOS POR GÉNERO

Blues

```python
import csv

# Parámetros de búsqueda
genre = "blues"
start_year, end_year = 2020, 2024
query = f"genre:{genre} year:{start_year}-{end_year}"

limit = 50              # máximo por llamada
max_total = 500         # nuevo límite total
offset = 0

blues_tracks_info = []

while True:
    results = sp.search(q=query, type='track', limit=limit, offset=offset)
    items = results.get('tracks', {}).get('items', [])
    if not items:
        break

    for item in items:
        track_name = item['name']
        album_name = item['album']['name']
        release_year = item['album'].get('release_date', '')[:4]  # solo año
        popularity = item['popularity']
        artists = ", ".join([artist['name'] for artist in item['artists']])
        blues_tracks_info.append({
            "track": track_name,
            "album": album_name,
            "release_year": release_year,
            "artists": artists,
            "popularity": popularity,
            "genre": genre,
            "year_range": f"{start_year}-{end_year}"
        })

    offset += limit
    if offset >= max_total:
        break

# Exportar a CSV
csv_filename = f"tracks_{genre}_{start_year}_{end_year}.csv"
fieldnames = ["track_id", "track", "album", "release_year", "artists", "popularity", "genre", "year_range"]

with open(csv_filename, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(blues_tracks_info)

# Mostrar un resumen en consola
for track in blues_tracks_info[:5]:
    print(track)
print(f"\nTotal tracks recibidos: {len(blues_tracks_info)}")
print(f"CSV guardado en: {csv_filename}")
```
